```python
import pandas as pd 
import numpy as np
```

```python
raw = pd.read_csv('./IR_WS_Dist.csv')
```

```python
raw['period_end'] = pd.to_datetime(raw['period_end'])
raw['month_name'] = raw['period_end'].dt.strftime('%B')
raw.head()
```

```python
factor_set = raw[['month_name','ghi','wind_speed_100m']]
factor_set.head()
```

```python
# Take 1000 random samples grouped by month
random_samples = factor_set.groupby('month_name').apply(lambda x: x.sample(1000, replace=False))

# Display the resulting DataFrame
random_samples.head()
```

```python
random_samples.reset_index(level='month_name', drop=True)
```

```python
grouped_samples = random_samples.groupby('month_name').apply(lambda x: pd.DataFrame({
    'ghi': x['ghi'],
    'ghi_probability': x['ghi'] / x['ghi'].sum(),
    'wind_speed_100m': x['wind_speed_100m'],
    'wind_speed_probability': x['wind_speed_100m'] / x['wind_speed_100m'].sum()
}))
```
